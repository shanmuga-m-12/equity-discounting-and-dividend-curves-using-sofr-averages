```python
import pandas as pd
import numpy as np

from scipy.stats import norm
```

```python
def bs_price_df(S0, K, T, sigma, DF_r, DF_d, option="call"):
    F = S0 * DF_r / DF_d
    vol_sqrt = sigma * np.sqrt(T)
    
    d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / vol_sqrt
    d2 = d1 - vol_sqrt
    
    if option == "call":
        price = DF_r * (F * norm.cdf(d1) - K * norm.cdf(d2))
    elif option == "put":
        price = DF_r * (K * norm.cdf(-d2) - F * norm.cdf(-d1))
    else:
        raise ValueError("option must be 'call' or 'put'")
        
    return price
```
